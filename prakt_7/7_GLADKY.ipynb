```python
from scipy import stats

table = [[3.83, 4.10, 3.63, 3.27, 14.83], [3.0, 3.13, 3.6, 3.5, 13.23], [2.7, 3.5, 3.2, 3.7, 13.1], [9.53, 10.73, 10.43, 10.47]]
# table = [[3,6,8,4,6,27], [7,5,6,7,2,27],[3,7,3,7,6,26],[6,4,2,8,6,26],[6,9,7,6,8,36],[7,4,8,4,9,32],[6,3,6,5,7,27],[3,2,9,8,6,28], [8,7,3,4,8,30],[3,8,8,7,1,27], [52,55,60,60,59]]
# данные с задания на паре
# правильный ответ на это задание: "S0 5.6","SnA 2.05","SnB 1.27", "S 4.61"


k = 3 # кол-во столбцов
m = 4 # кол-во строк
# n = 12
# 0.15137499999999984



x_i_avers = []
x_j_avers = []
t_summ = 0

for i in range(k):
    x_i_aver = table[i][-1]/m
    x_i_avers.append(x_i_aver)

for j in range(m):
    x_j_aver = table[-1][j]/k
    x_j_avers.append(x_j_aver)

for q in range(len(table)-1):
    t_summ += sum(table[q][:4])
x_aver = t_summ/12

summ1 = 0
summ2 = 0
summ3 = 0
for i in range(k):
    summ3 += (x_i_avers[i]-x_aver)**2
    for j in range(m):
        summ1 += (table[i][j] - x_i_avers[i])**2
        if i == 0:
            summ2 += (x_j_avers[j] - x_aver)**2

# k = 4
# m = 5

s_0 = (1/((k-1)*(m-1))) * (summ1 - k*summ2)
s_nA = (m/(k-1)) * summ3
s_nB = (k/(m-1)) * summ2

F_A = s_nA/s_0 # Чем больше F, тем значимее
F_B = s_nB/s_0

df1_A = k-1
df1_B = m-1
df2 = (k-1)*(m-1)


p_value_A = stats.f.sf(F_A, df1_A, df2)
p_value_B = stats.f.sf(F_B, df1_B, df2)  

# print(F_A, F_B, p_value_A, p_value_B, df1_A, df2)
# print("S0 5.6","SnA 2.05", "0.36", "SnB 1.27", "0.22", "S 4.61")

# Уровень значимости
alpha = 0.05
message_A = ""
message_B = ""
s = 0
s_A = 0
s_B = 0

# Сопоставляем S_nA S_nB и S_0 по критерию Фишера

# A
if p_value_A < alpha:
    s_A = (s_nA - s_0)/m
    message_A = "значимый"
else:
    s = ((k-1)*s_nA+(k-1)*(m-1)*s_0)/((k-1)+(k-1)*(m-1))
    message_A = "незначимый"

# B
if p_value_B < alpha:
    s_B = (s_nB - s_0)/k
    message_B = "значимый"
else:
    s = ((m-1)*s_nB+(k-1)*(m-1)*s_0)/((m-1)+(k-1)*(m-1))
    message_B = "незначимый"
   
if message_A == "незначимый" and message_B == "незначимый":
    s = ((k-1)*s_nA+(m-1)*s_nB+(k-1)*(m-1)*s_0)/((k-1)+(m-1)+(k-1)*(m-1))

print("S_0:", s_0)
print("S_nA:", s_nA)
print("S_nB:", s_nB)
print("Задание на паре:", "S0 5.6","SnA 2.05","SnB 1.27", "S 4.61")
if s_A != 0:
    print("Оценка фактора А:", s_A)
    print(s_A)
if s_B != 0:
    print("ОЦенка фактора B:", s_B)
    print(s_B)
if s != 0:
     print(f"Фактор А {message_A} и фактор В {message_B}. Оценка генеральной дисперсии: {s}")
# print(p_value_A)
# print(p_value_B)
5.62
```
