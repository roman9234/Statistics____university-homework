```python
import matplotlib.pyplot as plt
import math
import numpy as np

# Количество чисел в последовательности
n = 100

# Функция для генерации экспоненциально распределенных чисел
def generate_exponential_data(n, lambda_val):
    """
    n - кол-во чисел, которые генерируются
    lambda_val - интенсивность (или скорость) экспоненциального распределения. Чем больше значение lambda_val, тем быстрее убывает функция плотности вероятности
    """

    # uniform(0, 1, n) - генерирует массив из n случайных чисел, равномерного распределения на интервале [0, 1)
    # 1 - uniform(0, 1, n) - (0, 1]

    # np.log(...) -  применяется натуральный логарифм к каждому элементу массива. Логарифм используется для преобразования равномерного распределения в экспоненциальное

    # / lambda_val - мы делим результат на lambda_val, чтобы отмасштабировать сгенерированные данные в соответствии с интенсивностью экспоненциального распределения
    _data = -np.log(1 - np.random.uniform(0, 1, n)) / lambda_val
    filtered_arr = _data[_data <= 600]
    return filtered_arr

# Функции вычисление линейных коэффициентов корреляции
def get_b0(_x_values, _y_values):
    _n = len(_x_values)
    x_squared_sum = sum(list(x ** 2 for x in _x_values))
    x_sum = sum(_x_values)
    y_sum = sum(_y_values)
    y_x_sum = sum(list(_x_values[i] * _y_values[i] for i in range(_n)))

    # b0
    return (y_sum * x_squared_sum - y_x_sum * x_sum) / (_n * x_squared_sum - x_sum*x_sum)

def get_b1(_x_values, _y_values):
    _n = len(_x_values)
    x_squared_sum = sum(list(x ** 2 for x in _x_values))
    x_sum = sum(_x_values)
    y_sum = sum(_y_values)
    y_x_sum = sum(list(_x_values[i] * _y_values[i] for i in range(_n)))

    # b1
    return (y_sum*x_sum - y_x_sum * _n) / (x_sum*x_sum - _n*x_squared_sum)



def relative_frequencies(numbers: [int], min, max, bins):
    frequencies = []
    area = (max - min) / bins
    area_borders = [area * i for i in range(bins + 1)]
    counter = 0
    summ = 0
    for i in range(bins):
        for j in numbers:
            if area_borders[i] <= j < area_borders[i+1]:
                counter += 1
                summ += j
        # if counter == 0:
        #     counter += 0.1
        #     print("sdfsdfsdf")
        # frequencies.append(counter/(len(numbers) * area))
        frequencies.append(counter/(len(numbers)))
        counter = 0

    # Задние 2
    area_centers = [i + area/2 for i in area_borders[: -1]]
    plt.show()
    S_gist = sum([i for i in frequencies])
    print("Площадь гистограммы: ", S_gist)

    return frequencies, area, area_centers


# Функция для построения экспоненциальной аппроксимации
def plot_exponential_approximation(_t, _A, _b1):
    # _t - список значений
    # _A - амплитуда (или начальное значение) экспоненциальной функции. Это значение определяет, насколько высоко начинается функция
    # _b1 - коэффициент затухания. Этот параметр определяет скорость, с которой функция убывает со временем

    # вычисляет экспоненту (e в степени x) для каждого элемента в массиве -_b1 * _t
    f_t = _A * np.exp(-_b1 * _t)
    # Строим график _t и f(_t)
    plt.plot(_t, f_t, label="Экспоненциальная аппроксимация", color="red")




# Показательный закон:
# F(t) = A * e^(-b1*t)
# t  — случайная величина (например, время ожидания),
# b1 = λ > 0  — параметр интенсивности (обратная величина среднего времени ожидания).

# Математическое ожидание:  M = 1/λ
# Дисперсия D = 1/λ²

# ln(F(f)) = b0 - b1*t

# b1 - интенсивность величины случайного потока



lambda_val = 0.01  # Интенсивность
# Размеры последовательности для заданий 6 и 7
n_values = [100, 300, 500, 1000]
# n_values = [100]
errors = []  # Список для хранения погрешностей

for n in n_values:
    print(f"\n---- Выполнение заданий для длин последовательностей равных: n = {n} ----")

    # Задание 1 - Генерация данных
    data = generate_exponential_data(n, lambda_val)
    # for x in data:
    #     print(float(x))

    # Задание 2 - Построение гистограммы
    densities, area, centers = relative_frequencies(data, bins=10, min=0, max=600)

    print("\nЗадание 3 - Получение координат опорных точек")
    # Задание 3 - Получение координат опорных точек
    print("Опорные точки (центры интервалов и плотности):")
    for center, density in zip(centers, densities):
        print(f"x: {center:.2f}, y: {density:.5f}")

    print("\nЗадание 4 - Вычисление коэффициентов линейной регрессии, погрешности коэффициента интенсивности потока")
    # Задание 4 - Вычисление коэффициентов линейной регрессии

    # Для нахождения b0 и b1 можно использовать метод регресси для линейной однофакторной функции
    # Для этого в качестве опорных точек берём не значения плотности распределения времён, а их логорифмы.
    # Логорифм:

    x_values = centers
    y_values = [math.log(y) if y > 0 else math.log(0.00001) for y in densities]

    b0 = get_b0(x_values, y_values)
    b1 = get_b1(x_values, y_values)
    print(f"Коэффициенты линейной регрессии: b0 = {b0:.5f}, b1 = {b1:.5f}")

    # Погрешность:
    theoretical_lambda = lambda_val
    calculated_lambda = -b1
    # Вычисление относительной погрешности:
    error = abs(theoretical_lambda - calculated_lambda) / theoretical_lambda
    errors.append(error)
    print(f"Погрешность коэффициента интенсивности: {error:.2%}")

    print("\nЗадание 5 - Построение графика плотности обработки потсупающих программ от времени")
    # Задание 5 - Построение графика плотности обработки потсупающих программ
    plt.bar(centers, densities, area, edgecolor="k")
    plt.xlabel("Время обработки заявок (xi)")
    plt.ylabel("Плотность вероятности (pi)")
    plt.title(f"Гистограмма (n={n})")
    # plt.legend()
    plt.grid()
    plt.show()


    # Экспоненциальная аппроксимация
    A = np.exp(b0)
    # график экспоненциальной функции от 0 до 100 с вычисленным значением затухания
    t = np.linspace(0, max(data), 100)
    plot_exponential_approximation(t, A, -b1)
    plt.scatter(centers, densities, color="blue", label="Опорные точки")
    plt.xlabel("Времена обработки заявок")
    plt.ylabel("Плотность распределения случайной величины")
    plt.grid()
    plt.title("Плотность распределения случайной величины в зависимости от времени обработки заявок в исходных координатах")
    plt.legend()
    plt.show()



print("\n Задание 7 - График зависимости погрешности от размера последовательности")
# Задание 7 - График зависимости погрешности от размера последовательности
plt.figure(figsize=(8, 6))
plt.plot(n_values, errors, marker="o", linestyle="-", color="blue")
plt.xlabel("Размер последовательности (n)")
plt.ylabel("Погрешность")
plt.title("Зависимость погрешности от размера последовательности")
plt.ylim(0)
plt.grid()
plt.show()




```
